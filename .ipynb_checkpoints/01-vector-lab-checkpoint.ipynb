```python
import numpy as np
import matplotlib.pyplot as plt

v = np.array([1, 1])
a = 3
av = a * v

fig, ax = plt.subplots()
ax.plot(0, 0, 'or')
ax.arrow(0, 0, v[0], v[1], color='b', linewidth=2, head_width=0.2, head_length=0.25)
ax.arrow(0, 0, av[0], av[1], color='c', linestyle='dotted', linewidth=2, head_width=0.2, head_length=0.25)

plt.xlim(-2, 4)
plt.ylim(-1, 4)

plt.grid(True)

plt.show()
```
